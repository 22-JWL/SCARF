```python
"""
generate_augment.py
부족한 intent 발화 증강 → utterance_augment.csv 생성

대상 intent:
  - calibration_control  (목표 150개)
  - history_control      (목표 150개)
  - open_window          (목표 150개)
"""
# 3093
import random
import csv
from datetime import datetime, timedelta

random.seed(42)

# ────────────────────────────────────────────────────────────────
# calibration_control
# ────────────────────────────────────────────────────────────────

def generate_calibration(total=150):
    action_types    = ["button", "shape_similarity", "reference_select", "reticle_type", "camera"]

    button_kor = {
        "Test":      ["Test", "테스트"],
        "LightSave": ["LightSave", "라이트세이브", "조명 저장"],
    }
    shape_kor = {
        "rectangle": ["rectangle", "사각형", "직사각형"],
        "circle":    ["circle", "원", "동그라미"],
    }
    reference_kor = {
        "MULTIOBJECT": ["MULTIOBJECT", "멀티오브젝트", "다중 객체"],
        "CENTER":      ["CENTER", "센터", "중앙"],
        "BIGGEST":     ["BIGGEST", "비기스트", "가장 큰 객체"],
    }
    reticle_kor = {
        "NONE":     ["NONE", "논", "없음"],
        "DEFAULT":  ["DEFAULT", "디폴트", "기본"],
        "FULLSIZE": ["FULLSIZE", "풀사이즈", "전체 크기"],
    }
    camera_kor = {
        "Mapping":    ["Mapping", "매핑"],
        "SettingX1":  ["SettingX1", "세팅엑스원", "셋팅x1"],
        "SettingX2":  ["SettingX2", "세팅엑스투", "셋팅x2"],
        "PRS":        ["PRS", "피알에스"],
        "BarCode":    ["BarCode", "바코드"],
        "TopBarCode": ["TopBarCode", "탑바코드"],
        "Side":       ["Side", "사이드"],
    }

    def express(kor_map, key):
        return random.choice(kor_map[key])

    button_templates    = ["{b} 버튼 실행", "{b} 실행해줘", "{b} 눌러줘", "{b} 동작 시작", "{b} 버튼 눌러"]
    shape_templates     = ["{s} 모양 유사도 {v}로 설정", "{s} similarity {v}", "{s} 유사도 {v}%",
                           "{s} 유사도 {v}", "모양 {s} 유사도 {v}로"]
    reference_templates = ["기준 {r}로 설정", "{r} 기준 선택", "기준 {r}로 변경해줘", "{r}으로 기준 잡아",
                           "{r} 기준으로 변경"]
    reticle_templates   = ["십자선 {r}로 설정", "{r} reticle 설정", "십자선 타입 {r}로 변경",
                           "{r}로 십자선 바꿔", "십자선 {r}로 바꿔줘", "reticle {r}로 설정"]
    camera_templates    = ["{c} 카메라로 설정", "카메라 {c}로 변경", "{c}로 카메라 바꿔줘",
                           "카메라 {c} 선택", "{c} 카메라 설정"]
    none_templates      = ["캘리브레이션 설정해줘", "calibration 변경", "보정 설정", "캘리 설정",
                           "보정 다시 해줘", "캘리브레이션 옵션 조정", "캘리창 설정", "보정 파라미터 변경",
                           "보정 화면 설정", "캘리 파라미터 조정"]

    rows = []
    while len(rows) < total:
        r = random.random()
        action = random.choice(action_types)

        if r < 0.65:  # FULL
            if action == "button":
                key = random.choice(list(button_kor.keys()))
                b = express(button_kor, key)
                utt = random.choice(button_templates).format(b=b)
            elif action == "shape_similarity":
                key = random.choice(list(shape_kor.keys()))
                s = express(shape_kor, key)
                v = str(random.randint(0, 100))
                utt = random.choice(shape_templates).format(s=s, v=v)
            elif action == "reference_select":
                key = random.choice(list(reference_kor.keys()))
                rt = express(reference_kor, key)
                utt = random.choice(reference_templates).format(r=rt)
            elif action == "reticle_type":
                key = random.choice(list(reticle_kor.keys()))
                rt = express(reticle_kor, key)
                utt = random.choice(reticle_templates).format(r=rt)
            else:  # camera
                key = random.choice(list(camera_kor.keys()))
                c = express(camera_kor, key)
                utt = random.choice(camera_templates).format(c=c)

        elif r < 0.85:  # TYPE ONLY
            if action == "button":
                key = random.choice(list(button_kor.keys()))
                utt = f"{express(button_kor, key)} 버튼"
            elif action == "shape_similarity":
                key = random.choice(list(shape_kor.keys()))
                utt = f"{express(shape_kor, key)} 유사도 설정"
            elif action == "reference_select":
                utt = "기준 선택"
            elif action == "reticle_type":
                utt = "십자선 타입 설정"
            else:
                key = random.choice(list(camera_kor.keys()))
                utt = f"{express(camera_kor, key)} 카메라"

        else:  # NONE
            utt = random.choice(none_templates)

        rows.append(utt)

    return rows


# ────────────────────────────────────────────────────────────────
# history_control
# ────────────────────────────────────────────────────────────────

def generate_history(total=150):
    filter_types     = ["date", "camera", "inspection", "button"]
    camera_types     = ["NotSelected", "Mapping", "SettingX1", "SettingX2", "PRS", "BarCode", "TopBarCode", "Side"]
    inspection_types = ["Mark", "Qfn", "Bga", "Lga", "DataCode", "BottomDataCode", "Strip"]
    button_actions   = ["save", "open"]

    kor_map = {
        "Mapping": "매핑", "SettingX1": "세팅엑스원", "SettingX2": "세팅엑스투",
        "PRS": "피알에스", "BarCode": "바코드", "TopBarCode": "탑바코드",
        "Side": "사이드", "NotSelected": "미선택",
        "Mark": "마크", "Qfn": "큐에프엔", "Bga": "비지에이", "Lga": "엘지에이",
        "DataCode": "데이터코드", "BottomDataCode": "바텀데이터코드", "Strip": "스트립",
        "save": "저장", "open": "열기",
    }

    def express(v):
        return random.choice([v, kor_map.get(v, v)])

    def random_date():
        s = datetime(2024, 1, 1) + timedelta(days=random.randint(0, 365))
        e = s + timedelta(days=random.randint(1, 30))
        return f"{s.strftime('%Y-%m-%d')}_{e.strftime('%Y-%m-%d')}"

    date_templates       = ["날짜 {v}로 조회", "기록 기간 {v}", "{v} 기간 검색", "{v} 이력 보여줘"]
    camera_templates     = ["{v} 카메라 기록 조회", "camera {v}로 필터", "{v} 카메라 이력"]
    inspection_templates = ["{v} 검사 조회", "inspection {v} 선택", "{v} 검사 이력"]
    button_templates     = ["{v} 실행", "{v} 버튼 눌러", "{v} 동작"]
    none_templates       = ["이력 조회해줘", "history 검색", "검사 필터링", "기록 조회",
                            "이력 보여줘", "검사 기록 확인", "history 조회", "기록 필터 설정"]

    rows = []
    while len(rows) < total:
        r = random.random()
        ftype = random.choice(filter_types)

        if r < 0.6:  # FULL
            if ftype == "date":
                v = random_date()
                utt = random.choice(date_templates).format(v=v)
            elif ftype == "camera":
                v = express(random.choice(camera_types))
                utt = random.choice(camera_templates).format(v=v)
            elif ftype == "inspection":
                v = express(random.choice(inspection_types))
                utt = random.choice(inspection_templates).format(v=v)
            else:
                v = express(random.choice(button_actions))
                utt = random.choice(button_templates).format(v=v)

        elif r < 0.8:  # TYPE ONLY
            spoken = express(ftype)
            utt = f"{spoken} 필터 설정"

        elif r < 0.95:  # VALUE ONLY
            if random.random() < 0.5:
                v = express(random.choice(camera_types))
                utt = f"{v} 기록 조회"
            else:
                v = express(random.choice(button_actions))
                utt = f"{v} 실행"

        else:  # NONE
            utt = random.choice(none_templates)

        rows.append(utt)

    return rows


# ────────────────────────────────────────────────────────────────
# open_window
# ────────────────────────────────────────────────────────────────

def generate_open_window(total=150):
    window_names = ["lga", "qfn", "bga", "mapping", "strip",
                    "history", "light", "calibration", "setting", "lot"]
    kor_map = {
        "lga": "엘지에이", "qfn": "큐에프엔", "bga": "비지에이",
        "mapping": "매핑", "strip": "스트립", "history": "히스토리",
        "light": "조명", "calibration": "캘리브레이션", "setting": "설정", "lot": "랏"
    }
    templates     = ["{v} 창 열어", "{v} 열어줘", "{v} window open",
                     "{v} 창 켜", "{v} open", "{v} 창 띄워줘",
                     "{v} 화면 열어", "{v} 열어", "{v} 창 보여줘"]
    none_templates = ["창 열어줘", "window open", "화면 켜", "창 띄워줘",
                      "작업 창 열어", "티칭 화면 열어", "화면 보여줘"]

    rows = []
    while len(rows) < total:
        r = random.random()
        if r < 0.85:  # FULL
            w = random.choice(window_names)
            spoken = random.choice([w, kor_map[w]])
            utt = random.choice(templates).format(v=spoken)
        else:  # NONE
            utt = random.choice(none_templates)
        rows.append(utt)

    return rows


# ────────────────────────────────────────────────────────────────
# 저장
# ────────────────────────────────────────────────────────────────

if __name__ == "__main__":
    results = []

    for utt in generate_calibration(150):
        results.append({"utterance": utt, "intent": "calibration_control"})

    for utt in generate_history(150):
        results.append({"utterance": utt, "intent": "history_control"})

    for utt in generate_open_window(150):
        results.append({"utterance": utt, "intent": "open_window"})

    # 중복 제거
    seen = set()
    deduped = []
    for r in results:
        key = (r["utterance"], r["intent"])
        if key not in seen:
            seen.add(key)
            deduped.append(r)

    with open("cali_history_intent.csv", "a", newline="", encoding="utf-8-sig") as f:
        writer = csv.DictWriter(f, fieldnames=["utterance", "intent"])
        writer.writeheader()
        writer.writerows(deduped)

    # intent별 개수 출력
    from collections import Counter
    cnt = Counter(r["intent"] for r in deduped)
    print(f"총 {len(deduped)}개 생성")
    for k, v in cnt.items():
        print(f"  {k}: {v}개")
```
