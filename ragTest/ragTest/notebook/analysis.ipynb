```python
import json
from collections import defaultdict, Counter

# action.json 로드
with open("action.json", "r", encoding="utf-8") as f:
    actions = json.load(f)

# description → category 매핑
desc_to_cat = defaultdict(list)
for action in actions:
    desc = action.get("description")
    cat = action.get("category")
    if desc and cat:
        desc_to_cat[desc].append(cat)

print("Description, Category, Count")
for i, (desc, cats) in enumerate(desc_to_cat.items(), start=1):
    count = len(cats)
    unique_cats = ", ".join(set(cats))
    print(f"{i}. {desc} - {count}개 - Category: {unique_cats}")
```

```python
import pandas as pd

# CSV 불러오기
df = pd.read_csv("[cpu]label+rag_model_evaluation_detail.csv", encoding="utf-8-sig")

# 모델 리스트
model_list = df['Model'].unique()

# 결과 저장할 dict
most_common_wrong = {}
avg_wrong_counts = {}

for model in model_list:
    model_df = df[df['Model'] == model]
    
    # index별 등장 횟수 (무관 제외)
    counts = model_df['index'].value_counts()
    counts_no_irrelevant = counts.drop('irrelevant', errors='ignore')
    
    # 가장 많이 틀린 항목
    if not counts_no_irrelevant.empty:
        most_common = counts_no_irrelevant.idxmax()
        most_common_count = counts_no_irrelevant.max()
    else:
        most_common = None
        most_common_count = 0
    
    # 평균 틀린 개수 (무관 제외)
    avg_count = counts_no_irrelevant.mean() if not counts_no_irrelevant.empty else 0
    
    most_common_wrong[model] = (most_common, most_common_count)
    avg_wrong_counts[model] = avg_count

# 결과 출력
print("=== 모델별 가장 많이 틀린 항목 (무관 제외) 및 평균 ===")
for model in model_list:
    item, count = most_common_wrong[model]
    avg = avg_wrong_counts[model]
    print(f"{model}: 가장 많이 틀린 = {item} ({count}회), 평균 = {avg:.2f}회")
```

```python
import pandas as pd

# 예시 데이터
data = {
    "Model": ["ko-sroberta", "KoE5", "BGE-M3-ko", "KURE-v1"],
    "일반": [17, 25, 28, 29],
    "복합": [10, 13, 11, 9],
    "오타": [8, 13, 10, 10],
    "무관": [0, 0, 0, 0],
    "맞은 갯수": [35, 51, 49, 48],
}

# 전체 문항 수
total_dict = {"일반": 40, "복합": 20, "오타": 20, "무관": 20, "맞은 갯수": 100}

df = pd.DataFrame(data)

# 퍼센티지 계산
for col in ["일반", "복합", "오타", "무관", "맞은 갯수"]:
    df[col] = df[col] / total_dict[col] * 100

# 소수점 1자리로 포맷팅
df = df.round(1)

print(df)
```
